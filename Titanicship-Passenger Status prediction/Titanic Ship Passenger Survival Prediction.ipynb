```python
df.drop(["Ticket","Name","Cabin","PassengerId"], axis=1,inplace=True)
```

```python
X = df[['Age', 'Embarked', 'Fare','Parch','Pclass','Sex','SibSp','Title','Family_Size']] 
```

```python
X.shape
```

```python
y = df['Survived']
```

```python
y
```

```python
X = X.values
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
one = OneHotEncoder(sparse=False,handle_unknown='ignore')
b = one.fit_transform(X)
```

```python
b = b[:, 1:]
```

```python
b
```

```python
y = y.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(b, y,test_size=0.2, random_state=42)
```

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(X_val)
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_val, y_pred))
print('\n')
print(classification_report(y_val, y_pred))
```

```python
classifier.score(X_val,y_val)
```
